## Converting categorical data to numerical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab = LabelEncoder()
```

```python
df['T'] = lab.fit_transform(df['T'])
```

```python
df
```

```python
x = df.drop(["T"], axis=1)
```

```python
x
```

```python
y= df["T"]
```

```python
y
```

```python
sns.countplot(x=y)
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
```

```python
x_scale = sc.fit_transform(x)
x_scale
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x_scale, y,random_state=42, test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
y_train= knn.predict(x_test)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test, y_train))
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation="relu", input_shape=(x_train.shape[1],)),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(1, activation="linear")  # Linear activation for regression
])

# Compile the model
model.compile(optimizer="adam", loss="mean_squared_error")

# Training the model
model.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_test, y_test))
```

```python
loss = model.evaluate(x_test, y_test)
```

```python
print(f"MSE of test data = {loss}")
```
